#### 2.5.2 Removendo stop words e pontuação <a name="n3_removendo_stop_words"></a>

```python
stop_words_add = ['voce','vai','pra','de','pro','sao','vc','ta','essa', 'so', 'não', 'nao','seu','seus','sua','suas','eu','você', 'vocês','sou','somos','meu','minha','meus','minhas','teu','tua','teus','tuas','nosso','nossa','nossas','nossos', 'vosso','vossa','vossos','vossas', 'ja', 'ser', 'ai']
stop_words = nltk.corpus.stopwords.words('portuguese') + stop_words_add

processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = space_token.tokenize(sentence)
    
    for word in words:
        if word not in stop_words:
            new_sentence.append(word)
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.3 Pontuações <a name="n3_pontuacao"></a>

```python
punctuation_token = tokenize.WordPunctTokenizer()
list_punctuation = list()

for point in punctuation:
    list_punctuation.append(point)

punctuation_stopwords = list_punctuation + stop_words
processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in punctuation_stopwords:
            new_sentence.append(word)
            
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.4 Normalização do texto <a name="n3_normalizacao_texto"></a>

```python
without_accents = [unidecode.unidecode(word) for word in df.sentence]
without_accents_stop_words = [unidecode.unidecode(word) for word in punctuation_stopwords]

df.sentence = without_accents

processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in without_accents_stop_words:
            new_sentence.append(word)
    
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.5 Caracteres Maiúsculos <a name="n3_caracteres_maiusculo"></a>

```python
processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    sentence = sentence.lower()
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in without_accents_stop_words:
            new_sentence.append(word)
    
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.6 Stemming <a name="n3_stemming"></a>

```python
nltk.download('rslp')

stemmer = nltk.RSLPStemmer()
processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in without_accents_stop_words:
            new_sentence.append(stemmer.stem(word))
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

### 2.6. Classificadores <a name="n2_classificadores"></a>

#### 2.6.1 TF-IDF / Balanceamento <a name="n3_tf_idf_balanceamento"></a>

```python
# Para esse ponto, considera-se que as técnicas de pré-processamento foram realizadas. Aqui, será utilizado o classificador
tfidf = TfidfVectorizer();

# 1.Separar os dados de X e y:
X = tfidf.fit_transform(df.sentence)
y = df.classification

X = SelectPercentile(f_classif).fit_transform(X, y)
```

```python
# 2. Executar o train_test_split com stratify:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42, stratify = y)
```

```python
# 3. Normalizar o desbalanceamento com o Smote: 
sm = SMOTE(random_state = 42, sampling_strategy = 'minority')
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

#### 2.6.2 Global Class <a name="n3_global_class"></a>

```python
# Função para mostrar os resultados
def plot_result(results):
    train_score = results['mean_train_score'] 
    test_score = results['mean_test_score']
    final = []
    
    for i in range(len(train_score)):
        final.append([i, train_score[i], test_score[i]])
    
    final = pd.DataFrame(final, columns= ['execute_number', 'mean_train_score', 'mean_test_score']) 
    print(final) 
    
    sns.lineplot(x = 'execute_number', y = 'mean_train_score', data = final)
    sns.lineplot(x = 'execute_number', y = 'mean_test_score', data = final)
    plt.ylim(0, 1)
    plt.legend(['Train', 'Test']) 
    

#Função para criar a classificação de acordo com parâmetro passado.
def make_classification(params,classifier, X_train_res, y_train_res, X_test, y_test, model_name, count = False):
    # Usar count=True para MultinomialNB
    # Para esse ponto, considera-se que as técnicas de pré-processamento foram realizadas. Aqui, será utilizado o classificador

    # 4. Processar um GridSearchCV para escolha da melhor configuração:
    if(params):
        search = GridSearchCV(classifier(), params, cv = 10, scoring = 'f1', return_train_score = True, n_jobs = -1)
    else:
        search = classifier()
    search.fit(X_train_res, y_train_res)
    if hasattr(search, 'cv_results_'):
        results = pd.DataFrame(search.cv_results_)

    # 5. Apresentar os resultados de acordo com as configurações:    
    y_pred = search.predict(X_test)  

    # 6. Obtendo a F Medida (verificar como calcular F Medida):
    f_measure = f1_score(y_test, y_pred, average = 'binary')
    print('F Medida: %.2f' % f_measure)
    print('--------------------------------------------')

    # 7. Apresentar a melhor configuração:
    if hasattr(search, 'best_estimator_'):
        print(search.best_estimator_)
        print('Score do GridSearchCV (melhor): ', search.best_score_)
        print('--------------------------------------------')
    
    if hasattr(search, 'cv_results_'):
        print('Médias das execuções:')
        plot_result(results)        
        print('--------------------------------------------')

    # 8. Relatórios:
    print('Classification Reports: ')
    print(classification_report(y_test, y_pred))
    print('--------------------------------------------')

    print('Matriz de Confusão: ')
    print(pd.crosstab(y_test, y_pred, rownames=['Real'], colnames=['Predito'], margins=True))
    print('--------------------------------------------')

    persist_model(search, model_name)

#Criação do arquivo
def persist_model(model, name):
    dump(model, name + '.gz')
```
