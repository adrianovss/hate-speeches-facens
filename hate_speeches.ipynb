## 1. Contextualização <a name="n1_contextualizacao"></a>

### 1.1. Problemas identificados <a name="n2_problemas_identificados"></a>

  Os sentimentos e opiniões através de textos, sendo positivo ou negativo, é uma grande utilidade para análise de dados utilizado em pesquisas e melhorias de determinado cenário.
  Até mesmo para os humanos, extrair e analisar opiniões e sentimentos de um texto não é uma tarefa fácil, com isso de forma inteligente e cada vez mais frequente as empresas aplicam algoritmos e modelos computacionais para a resolução desse tipo de problema.


### 1.2. Objetivos <a name="n2_objetivos"></a>

Este trabalho propõe a utilização tecnologias de aprendizado de máquina e processamento de linguagem natural mais precisamente a implementação  de um modelo para realizar a classificação de palavras de tipo ofensivo ou não ofensivo.

### 1.3. Dataset escolhido <a name="n2_dataset_escolhido"></a>

O dataset utilizado neste (**OFFCOMBR-2**) possui um conjunto de dados coletados dos portais web G1 Notícias - com ênfase em discursos relacionados com política e esportes. O artefato possui três colunas específicas:

- ID (numerando a sequência de sentenças presentes no corpus);
- Class (identificando a classificação como discurso ofensivo ou não);

#### 2.5.2 Removendo stop words <a name="n3_removendo_stop_words"></a>

```python
stop_words_add = ['voce','vai','pra','de','pro','sao','vc','ta','essa', 'so', 'não', 'nao','seu','seus','sua','suas','eu','você', 'vocês','sou','somos','meu','minha','meus','minhas','teu','tua','teus','tuas','nosso','nossa','nossas','nossos', 'vosso','vossa','vossos','vossas', 'ja', 'ser', 'ai']
stop_words = nltk.corpus.stopwords.words('portuguese') + stop_words_add

processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = space_token.tokenize(sentence)
    
    for word in words:
        if word not in stop_words:
            new_sentence.append(word)
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.3 Pontuações <a name="n3_pontuacao"></a>

```python
punctuation_token = tokenize.WordPunctTokenizer()
list_punctuation = list()

for point in punctuation:
    list_punctuation.append(point)

punctuation_stopwords = list_punctuation + stop_words
processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in punctuation_stopwords:
            new_sentence.append(word)
            
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.4 Normalização do texto <a name="n3_normalizacao_texto"></a>

```python
without_accents = [unidecode.unidecode(word) for word in df.sentence]
without_accents_stop_words = [unidecode.unidecode(word) for word in punctuation_stopwords]

df.sentence = without_accents

processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in without_accents_stop_words:
            new_sentence.append(word)
    
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.5 Caracteres maiúsculos <a name="n3_caracteres_maiusculo"></a>

```python
processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    sentence = sentence.lower()
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in without_accents_stop_words:
            new_sentence.append(word)
    
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.5.6 Stemming <a name="n3_stemming"></a>

```python
nltk.download('rslp')

stemmer = nltk.RSLPStemmer()
processed_sentence = list()

for sentence in df.sentence:
    new_sentence = list()
    
    words = punctuation_token.tokenize(sentence)
    
    for word in words:
        if word not in without_accents_stop_words:
            new_sentence.append(stemmer.stem(word))
    processed_sentence.append(' '.join(new_sentence))
    
df.sentence = processed_sentence
df.head()
```

#### 2.6.2 Funções globais <a name="n3_funcoes_globais"></a>

```python
class SVDDimSelect(object):
    def fit(self, X, y=None):        
        try:
            self.svd_transformer = TruncatedSVD(n_components=round(X.shape[1]/2))
            self.svd_transformer.fit(X)
        
            cummulative_variance = 0.0
            k = 0
            for var in sorted(self.svd_transformer.explained_variance_ratio_)[::-1]:
                cummulative_variance += var
                if cummulative_variance >= 0.5:
                    break
                else:
                    k += 1
                
            self.svd_transformer = TruncatedSVD(n_components=k)
        except Exception as ex:
            print(ex)
            
        return self.svd_transformer.fit(X)
    
    def transform(self, X, Y=None):
        return self.svd_transformer.transform(X)
        
    def get_params(self, deep=True):
        return {}
    
def adjust_data(dense_data = 0):
    X_train, X_test, y_train, y_test = train_test_split(df.sentence, df.classification, test_size = 0.3, random_state = 42, stratify = df.classification)

    tfidf = TfidfVectorizer()

    tf_idf = tfidf.fit(X_train, y_train)
    X_train = tf_idf.transform(X_train)
    X_test = tf_idf.transform(X_test)

    feature_selection = SelectPercentile(f_classif, percentile = 30).fit(X_train, y_train)

    X_train = feature_selection.transform(X_train)
    X_test = feature_selection.transform(X_test)

    sm = SMOTE(random_state = 42, sampling_strategy = 'minority')
    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
    
    if(dense_data == 1):
        X_train_res = X_train_res.toarray()
        X_test = X_test.toarray()
    
    return X_train_res, X_test, y_train_res, y_test
    
def view_overfitting(results):
    train_score = results['mean_train_score'] 
    test_score = results['mean_test_score']
    
    final = []
    
    for i in range(len(train_score)):
        final.append([i, train_score[i], test_score[i]])
    
    final = pd.DataFrame(final, columns= ['execute_number', 'mean_train_score', 'mean_test_score']) 
    
    sns.lineplot(x = 'execute_number', y = 'mean_train_score', data = final)
    sns.lineplot(x = 'execute_number', y = 'mean_test_score', data = final)
    plt.ylim(0, 1)
    plt.legend(['Train', 'Test']) 
    plt.title('View Overfitting')
    

def make_reports(search, X_test, y_test, y_pred, results):
    f_measure = f1_score(y_test, y_pred, average = 'binary')
    
    print('F Medida: %.2f \n' % f_measure)

    if hasattr(search, 'best_estimator_'):
        print('Melhor configuração: ', search.best_estimator_, '\n')
        print('Melhor Score (F1): ', search.best_score_, '\n')
    
    if hasattr(search, 'cv_results_'):
        view_overfitting(results)        
        print('\n')

    print('Classification Reports: ')
    print(classification_report(y_test, y_pred))
    print('\n')
    
    '''
        print('Matriz de Confusão: ')
        print(pd.crosstab(y_test, y_pred, rownames=['Real'], colnames=['Predito'], margins=True))
    '''
    
    disp = plot_confusion_matrix(search, X_test, y_test, values_format = 'd')
    disp.figure_.suptitle("Confusion Matrix")    
    
def persist_model(model, name):
    dump(model, name + '.gz')
    
def make_classification_primary(params, classifier, model_name, hyperparameter_executor = 'GS'):
    search = ''
    results = ''
    
    X_train, X_test, y_train, y_test = adjust_data() if model_name is not 'gaussianNaiveBayes' else adjust_data(1)
    
    if(hyperparameter_executor == 'GS'):
        search = GridSearchCV(classifier(), params, cv = 10, scoring = 'f1', return_train_score = True, n_jobs = -1) if params else classifier()
        
    elif(hyperparameter_executor == 'RS'):
        search = RandomizedSearchCV(classifier(), params, cv = 10, scoring = 'f1', return_train_score = True, n_jobs = -1) if params else classifier()
        
    search.fit(X_train, y_train)
    
    if hasattr(search, 'cv_results_'):
        results = pd.DataFrame(search.cv_results_)

    y_pred = search.predict(X_test)  

    make_reports(search, X_test, y_test, y_pred, results)
    persist_model(search, model_name)
    
    
def make_classification_secundary(params, classifier, model_name, hyperparameter_executor = 'GS'):
    search = ''
    results = ''
    
    X_train, X_test, y_train, y_test = train_test_split(
        df.sentence, df.classification, test_size = 0.3, random_state = 42, stratify = df.classification)
    
    tfidf = TfidfVectorizer()

    tf_idf = tfidf.fit(X_train, y_train)
    X_train = tf_idf.transform(X_train)
    X_test = tf_idf.transform(X_test)
    
    pipeline = Pipeline([
        ("smt", SMOTE(sampling_strategy = 'minority')),
        ("svd", SVDDimSelect()),
        ("clf", classifier())
    ])
    
    if(model_name == 'gaussianNaiveBayes'):
        search = pipeline
    
    else:
        if(hyperparameter_executor == 'GS'):
            search = GridSearchCV(pipeline, params, cv = 10, scoring = 'f1', return_train_score = True, n_jobs = -1) if params else classifier()
        
        elif(hyperparameter_executor == 'RS'):
            search = RandomizedSearchCV(pipeline, params, cv = 10, scoring = 'f1', return_train_score = True, n_jobs = -1) if params else classifier()
        
    search.fit(X_train, y_train)
    
    if hasattr(search, 'cv_results_'):
        results = pd.DataFrame(search.cv_results_)

    y_pred = search.predict(X_test)  

    make_reports(search, X_test, y_test, y_pred, results)
    persist_model(search, model_name)
```
