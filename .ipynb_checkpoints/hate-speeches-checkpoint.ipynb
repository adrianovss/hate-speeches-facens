#### 2.5.2 Definição de Classes <a name="n3_definicao_classes"></a>

```python
class PreProcessing(object):
    def wordTokenization(self, dataset, column):
        return [word_tokenize(sentence) for sentence in dataset[column]]
    
    def removeStopWords(self, sentences):
        stopwordsList = stopwords.words('portuguese')
        
        nonStopWords = []
        
        for sentence in sentences:
            nonStopWords.append([word for word in sentence if not word.lower() in stopwordsList])
            
        return nonStopWords
    
    def removePonctuactions(self, sentences):
        punctuations = string.punctuation
        
        nonPontuactions = []
        
        for sentence in sentences:
            nonPontuactions.append([word for word in sentence if not word in punctuations])
        
        return nonPontuactions
    
    def executeStemming(self, sentences):
        stemmer = SnowballStemmer('portuguese')
        stemmings = []
        
        for sentence in sentences:
            stemmings.append([stemmer.stem(word) for word in sentence])
        
        return stemmings
    
    def executeLemmatization(self, sentences):
        lemmatizer = WordNetLemmatizer()
        lemmatizations = []
        
        for sentence in sentences:
            posTags = nltk.pos_tag(sentence)
            lemmas = []
            
            for w in posTags:
                if w[1].startswith('J'):
                    pos_tag = wordnet.ADJ
                elif w[1].startswith('V'):
                    pos_tag = wordnet.VERB
                elif w[1].startswith('N'):
                    pos_tag = wordnet.NOUN
                elif w[1].startswith('R'):
                    pos_tag = wordnet.ADV
                else:
                    pos_tag = wordnet.NOUN

                lemmas.append(lemmatizer.lemmatize(w[0], pos_tag))

            lemmatizations.append(lemmas)
            
        return lemmatizations
```

```python
pre_processing = PreProcessing()
```

### 2.6. Modelo e Classificação <a name="n2_modelo_classificacao"></a>

#### 2.6.1 BOW (CBOW, Glove) / Word2Vec <a name="n3_bow_word_two_vec"></a>

```python
tokenizer = Tokenizer(num_words = None, lower = True, split = ' ', char_level = False, oov_token = None, document_count = 0)
words = tokenizer.fit_on_texts(non_pontuactions)
X1 = tokenizer.texts_to_sequences(non_pontuactions)

max_features = len(tokenizer.word_index)
print('Words Count: ' + str(max_features))

maxlen = max([len(sen) for sen in X1])
print("Max Sentence: " + str(maxlen))
```

```python
X1 = pad_sequences(X1, maxlen=maxlen)
Y1 = pd.get_dummies(df['classification']).values
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1,Y1, random_state = 42)

print(X1_train.shape,Y1_train.shape)
print(X1_test.shape,Y1_test.shape)
```

```python
print(X1)
print(Y1)
```

#### 2.6.2 LSTM <a name="n3_lstm"></a>

```python
embed_dim = 150
lstm_out = 200
dropout = 0.2
model = Sequential()
model.add(Embedding((max_features+1), embed_dim,input_length = X1.shape[1]))
model.add(SpatialDropout1D(dropout))
model.add(LSTM(lstm_out, dropout=dropout, recurrent_dropout=dropout))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])

print(model.summary())
```

```python
batch_size = 64
history = model.fit(X1_train, Y1_train, epochs = 10,shuffle=True, batch_size=batch_size, verbose = 2, validation_split=0.3, callbacks=[EarlyStopping(monitor='val_loss')])
```

```python
print("Loss: {}".format(str(history.history['loss'])))
print("Accuracy: {}".format(str(history.history['accuracy'])))
print("Validation Loss: {}".format(str(history.history['val_loss'])))
print("Validation Accuracy: {}".format(str(history.history['val_accuracy'])))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model train vs validation')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss train', 'loss validation', 'acc train', 'acc validation'], loc='upper left')
plt.show()
```

```python
print("Evaluate: {}".format(str(model.evaluate(X1_test, Y1_test, verbose=0))))

y_pred = model.predict(X1_test)
y_pred[:10]
```

```python
def plot_confusion_matrix_not_scikit(cm, class_names):
    """
    Returns a matplotlib figure containing the plotted confusion matrix.
    
    Args:
       cm (array, shape = [n, n]): a confusion matrix of integer classes
       class_names (array, shape = [n]): String names of the integer classes
    """
    
    figure = plt.figure(figsize=(8, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title("Confusion matrix")
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)
        
    # Use white text if squares are dark; otherwise black.
    threshold = cm.max() / 2.
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        color = "white" if cm[i, j] > threshold else "black"
        plt.text(j, i, cm[i, j], horizontalalignment="center", color=color)
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    return figure

print("Classification report for classifier LSTM:\n%s\n" % classification_report(Y1_test.argmax(axis=1), y_pred.argmax(axis=1)))
cm = confusion_matrix(Y1_test.argmax(axis=1), y_pred.argmax(axis=1))
fig = plot_confusion_matrix_not_scikit(cm, [0,1])
fig.show()
```
